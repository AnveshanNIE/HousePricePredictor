```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("house_prices.csv")
```

* One of the common way to analyze the data is through visualization

```python
plt.scatter(df["houseSize"], df["price"], color="blue", label="HousePrice")
plt.title('House Size vs. Price')
plt.xlabel('House Size (sqft)')
plt.ylabel('Price (Rs)')
plt.grid(True)
plt.legend()
plt.show()
```

### Linear regression training phase

* We will be using library called scikit-learn

        ! pip install sklearn

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Define data and target
data = df[["houseSize"]].values
target = df["price"].values

# Create Linear Regression model
lr_model = LinearRegression()

# Fit the model
print("Training the model")
lr_model.fit(data, target)

# Make predictions on the entire dataset
y_pred = lr_model.predict(data)
print("Predicted values:", y_pred)
```

* Checking the accuracy of the model

```python
from sklearn.metrics import r2_score
r2 = r2_score(target, y_pred)
print(f"R-squared: {r2:.2f}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Left Plot: Actual Data (houseSize vs price)
axes[0].scatter(df["houseSize"], df["price"], color="red", label="Actual Prices")
axes[0].set_title('Actual Data (houseSize vs Price)')
axes[0].set_xlabel('House Size (sqft)')
axes[0].set_ylabel('Price ($)')
axes[0].grid(True)
axes[0].legend()

# Right Plot: Predicted Data (houseSize vs y_pred)
axes[1].scatter(df["houseSize"], y_pred, color="blue", label="Predicted Prices")
axes[1].set_title('Predicted Data (houseSize vs Predicted Price)')
axes[1].set_xlabel('House Size (sqft)')
axes[1].set_ylabel('Price ($)')
axes[1].grid(True)
axes[1].legend()

plt.tight_layout()
plt.show()
```
